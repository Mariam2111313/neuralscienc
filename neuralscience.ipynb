```python
import numpy as np

def tanh(x):
    return np.tanh(x)

x = np.array([0.6, -0.3])   

w1 = np.random.uniform(-0.5, 0.5, size=2)  
w2 = np.random.uniform(-0.5, 0.5)           


b1 = 0.5
b2 = 0.7


z1 = np.dot(x, w1) + b1
a1 = tanh(z1)

z2 = a1 * w2 + b2
output = tanh(z2)

print("Network Output:", output)
```

```python
import numpy as np
import matplotlib.pyplot as plt


def tanh(x):
    return np.tanh(x)


b1 = 0.5
b2 = 0.7


w1 = np.random.uniform(-0.5, 0.5, size=2)  
w2 = np.random.uniform(-0.5, 0.5)        


x_values = np.linspace(-2, 2, 200)

outputs = []


for x in x_values:
    input_vector = np.array([x, x])  
    
    z1 = np.dot(input_vector, w1) + b1
    a1 = tanh(z1)
    
    z2 = a1 * w2 + b2
    output = tanh(z2)
    
    outputs.append(output)


plt.figure()
plt.plot(x_values, outputs)
plt.xlabel("Input Value")
plt.ylabel("Network Output")
plt.title("Neural Network Output using tanh Activation")
plt.show()
```
